```python
from pathlib import Path

from prepare_data import get_dataframe

crud = get_dataframe(
    Path("../../Static Analysis/impl-crud/crud_martin.csv"),
    application="CRUD"
)

cqrs = get_dataframe(
    Path("../../Static Analysis/impl-es-cqrs/es-cqrs-martin.csv"),
    application="ES CQRS"
)
```

```python
from colors import APP_COLORS
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.concat([crud, cqrs])
df = df.groupby(['A', 'I', 'Application']).size().reset_index(name='Count')

fig = plt.figure(figsize=(9, 9))

sns.set_theme(style="whitegrid")

sns.scatterplot(
    data=df,
    x="I",
    y="A",
    hue="Application",
    palette=APP_COLORS,
    size='Count',
    sizes=(40, 400)
)

plt.title("Main Sequence Graph (Martin Metrics)")
plt.xlabel("Instability")

plt.ylabel("Abstractness")
ax = fig.axes[0]

ax.set_ylim(-0.05, 1)
ax.set_xlim(-0.05, 1.05)
ax.plot([0, 1], [1, 0], color="gray")

plt.legend(
    labelspacing=1.2,
    borderpad=1.2,
    handletextpad=1.0,
    loc='upper right',
    bbox_to_anchor=(1.05, 1)
)

plt.show()
```
