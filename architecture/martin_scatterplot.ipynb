```python
from pathlib import Path

from prepare_data import get_dataframe

crud = get_dataframe(
    Path("../../Static Analysis/impl-crud/crud_martin.csv"),
    application="CRUD"
)

cqrs = get_dataframe(
    Path("../../Static Analysis/impl-es-cqrs/es-cqrs-martin.csv"),
    application="ES CQRS"
)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.concat([crud, cqrs])

fig = plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x="I",
    y="A",
    hue="Application",
)

plt.title("Main Sequence Graph (Martin Metrics)")
plt.xlabel("Instability")

plt.ylabel("Abstractness")
ax = fig.axes[0]

ax.plot([0, 1], [0, 1], transform=ax.transAxes, color="gray", alpha=.7)

plt.legend()

plt.show()
```
